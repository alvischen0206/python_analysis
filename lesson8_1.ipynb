```python
import pandas as pd
import numpy as np

# 1. Create a Sample DataFrame (No previous context assumed)
np.random.seed(0)
subjects = ['國文', '英文', '數學', '自然', '社會']
student_names = ['小明', '小華', '小美', '小強', '小玉']
scores = np.random.randint(50, 101, size=(len(student_names), len(subjects)))
df = pd.DataFrame(scores, columns=subjects, index=student_names)

# 2. Add a "總分" column
df['總分'] = df.sum(axis=1)

# 3. Sort by "總分" and add the "名次" column
df_sorted = df.sort_values(by='總分', ascending=False)
df_sorted['名次'] = df_sorted['總分'].rank(ascending=False, method='min').astype(int)

# 4. Move "名次" to the front
cols = ['名次'] + [col for col in df_sorted.columns if col != '名次']
df_reordered = df_sorted[cols]

# 5. Style the DataFrame (highlight low scores in red)
def highlight_low_scores(val):
    """Highlights scores below 60 in red."""
    color = 'red' if isinstance(val, (int, float)) and val < 60 else 'white'
    return f'color: {color}'

styled_df = df_reordered.style.applymap(highlight_low_scores)

# 6. Display the Styled DataFrame
print("最終成績表 (已對齊，低於 60 分標示為紅色，名次在最前面)：")
display(styled_df)
```
