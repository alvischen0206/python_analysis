```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.set_xticks([1, 3, 5, 7, 9])
axes.set_yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.plot([11, 3, 5, 7, 9],marker='o',color='red', linestyle='')
plt.show()
```

```python
#  2x2 
figure = plt.figure(figsize=(12, 8))

#  ()
ax1 = figure.add_subplot(2, 2, 1)
ax1.plot([11, 3, 5, 7, 9], marker='o', color='red', linestyle='')
ax1.set_title('')

#  ()
ax2 = figure.add_subplot(2, 2, 2)
ax2.bar(['a', 'b', 'c', 'd', 'e'], [1, 2, 3, 4, 5])
ax2.set_title('')

#  ()
ax3 = figure.add_subplot(2, 2, 3)
ax3.pie([30, 20, 25, 15, 10], 
        labels=['A', 'B', 'C', 'D', 'E'],
        autopct='%1.1f%%')
ax3.set_title('')

#  ()
ax4 = figure.add_subplot(2, 2, 4)
ax4.plot([1, 4, 2, 5, 3], marker='s', color='green')
ax4.set_title('')
ax4.annotate('', 
            xy=(2, 2),           #  ()
            xytext=(2.5, 2.5),     # 
            color='red',         # 
            arrowprops=dict(
                facecolor='red',    # 
                shrink=0.1,         # 
                connectionstyle='arc3,rad=-.3',  # rad
                width=2,            # 
                headwidth=10        # 
            ))
# 
plt.tight_layout()

# 
plt.show()
```
