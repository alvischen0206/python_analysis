```python
import pandas as pd

# 設定 Pandas 顯示選項
pd.set_option('display.unicode.ambiguous_as_wide', True)
pd.set_option('display.unicode.east_asian_width', True)

# 讀取 Excel 檔案
df = pd.read_excel('student_scores.xlsx', index_col='學號')

# 方法 1：使用 loc 搭配列表
print("指定學號的學生成績：")
print("-" * 50)
selected_students = df.loc[['A1143033017', 'A1143033033']]
print(selected_students)

# 方法 2：使用 iloc搭配列表
print("方法2")
print("-" * 50)
selected_students = df.iloc[[19,20]]
print(selected_students)

```

```python
# 設定 Pandas 顯示選項
pd.set_option('display.unicode.ambiguous_as_wide', True)
pd.set_option('display.unicode.east_asian_width', True)

# 讀取 CSV 檔案
df = pd.read_csv('student_scores.csv', index_col=0)

# 列出要處理的科目欄位
subjects = ['國文', '英文', '數學', '理化', '公民']



# 儲存修改後的成績
df.to_csv('student_scores_adjusted.csv', encoding='utf-8-sig')
print("\n已將調整後的成績儲存至 student_scores_adjusted.csv")

# 計算總分和平均
df['總分'] = df[subjects].sum(axis=1)
df['平均'] = df[subjects].mean(axis=1).round(2)

print("\n加入總分和平均後的成績表：")
print("-" * 50)
print(df)

# 儲存更新後的成績
df.to_csv('student_scores_final.csv', encoding='utf-8-sig')
print("\n已將加入總分和平均的成績表儲存至 student_scores_final.csv")
```
